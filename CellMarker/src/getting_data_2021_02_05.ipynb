Data at CellMarker is made available in txt files that follow the TSV format. 

I will import the datasets and save here. 

```python
import pandas as pd
```

```python
human_markers_2021_02_05 = pd.read_csv("http://bio-bigdata.hrbmu.edu.cn/CellMarker/download/Human_cell_markers.txt",
                                       sep="\t" )
```

```python
mouse_markers_2021_02_05 = pd.read_csv("http://bio-bigdata.hrbmu.edu.cn/CellMarker/download/Mouse_cell_markers.txt",
                                       sep="\t" )
```

```python
human_markers_2021_02_05.drop_duplicates().to_csv("./data/human_markers_2021_02_05.tsv", sep="\t", index=False)
```

```python
mouse_markers_2021_02_05.drop_duplicates().to_csv("./data/mouse_markers_2021_02_05.tsv", sep="\t", index=False)
```
