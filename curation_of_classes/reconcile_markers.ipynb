```python
def split_markers(marker_string):
    
    marker_string = marker_string.replace("+", " ")
    marker_string = marker_string.replace("(", " ")
    marker_string = marker_string.replace(")", " ")
    marker_string = marker_string.replace(")", " ")

    sep_markers = marker_string.split(",")
    sep_markers = [s.replace("and", "").strip() for s in sep_markers]
    return(sep_markers)
```
